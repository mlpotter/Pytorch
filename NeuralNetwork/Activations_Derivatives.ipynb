```python
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-10,10,steps=100,dtype=torch.float32,requires_grad=True)
```

## Sigmoid

```python
y = torch.sigmoid(x)
Y = y.sum()
Y.backward()
```

```python
plt.figure(figsize=(20,10))
plt.plot(x.detach().numpy(),y.detach().numpy(),linewidth=5,label="Sigmoid")
plt.plot(x.detach().numpy(),x.grad.numpy(),linewidth=5,label="$\dfrac{dy}{dx}$")
plt.legend(fontsize=20)
plt.xlabel("x",fontsize=30)
plt.ylabel("y",fontsize=30)
plt.title("Sigmoid",fontsize=30)
plt.show()

x.grad.zero_()
```
