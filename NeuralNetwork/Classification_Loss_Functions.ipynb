```python
import numpy as np

import matplotlib.pyplot as plt
```

## MSE with hard threshold function as loss function

```python
n_points_per_class = 4

x = np.concatenate((np.linspace(-5,-1,n_points_per_class),np.linspace(1,5,n_points_per_class)),0)
y = np.concatenate((np.zeros((n_points_per_class,)),np.ones((n_points_per_class,))),0)
decision_threshold =np.arange(-5,5,10/100)

losses = []
for threshold in decision_threshold:
    losses.append(np.mean((y - (x > threshold))**2))
```

```python
plt.figure(figsize=(15,10))
plt.plot(decision_threshold.tolist(),losses,label="Loss Function MSE",linewidth=5)
plt.plot(x.tolist()[0:n_points_per_class],[0]*n_points_per_class,'r*',label="Class 0",markersize=20)
plt.plot(x.tolist()[n_points_per_class:],[0]*n_points_per_class,'g*',label="Class 1",markersize=20)
plt.legend(fontsize=20)
plt.xlabel("x",fontsize=20)
plt.show()
```

## MSE with sigmoid threshold function as loss function

```python
losses = []
for threshold in decision_threshold:
    losses.append(np.mean((y - 1/(1+np.exp(-(x*4 + threshold))))**2))
```

```python
plt.figure(figsize=(15,10))
plt.plot(decision_threshold.tolist(),losses,label="Loss Function MSE",linewidth=5)
plt.plot(x.tolist()[0:n_points_per_class],[0]*n_points_per_class,'r*',label="Class 0",markersize=20)
plt.plot(x.tolist()[n_points_per_class:],[0]*n_points_per_class,'g*',label="Class 1",markersize=20)
plt.legend(fontsize=20)
plt.xlabel("x",fontsize=20)
plt.show()
```

## CE loss function

```python
losses = []
for threshold in decision_threshold:
    y_hat = 1/(1+np.exp(-(x*4 + threshold)))
    losses.append(-np.mean(y*np.log(y_hat) + (1-y)*np.log(1-y_hat)))
```

```python
plt.figure(figsize=(15,10))
plt.plot(decision_threshold.tolist(),losses,label="Loss Function CE",linewidth=5)
plt.plot(x.tolist()[0:n_points_per_class],[0]*n_points_per_class,'r*',label="Class 0",markersize=20)
plt.plot(x.tolist()[n_points_per_class:],[0]*n_points_per_class,'g*',label="Class 1",markersize=20)
plt.legend(fontsize=20)
plt.xlabel("x",fontsize=20)
plt.show()
```
